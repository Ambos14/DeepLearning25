# Multi-Class Perceptrons

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs #for data generatio
```

```python
X, y = make_blobs(n_samples=50, centers=3, cluster_std=0.2,n_features=2,center_box=(1,5),random_state=104)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.colorbar()
```

### draw Percepton layout
* Hint: now we have more then one Perceptron
* write math. formulation as linear model

```python
def plot_line_from_normal_vector_and_offset(normal_vector, offset, x_range=None, ax=None, **kwargs):
    """
    Plots a 2D line defined by its normal vector w and offset b such that wx + b = 0.

    Args:
        normal_vector (list or numpy.ndarray): A 2D vector representing the normal to the line (w). Must be of length 2.
        offset (float): The offset (b) such that w.x + b = 0.
        x_range (tuple, optional): A tuple (xmin, xmax) specifying the range of x-values to plot.
                                     If None (default), a reasonable range is automatically determined.
        ax (matplotlib.axes._axes.Axes, optional): The matplotlib axes object to plot on. If None (default),
                                                  the current axes (plt.gca()) is used, or a new figure and axes
                                                  are created.
        **kwargs: Keyword arguments to pass to the `plot` function (e.g., color, linestyle, linewidth).

    Returns:
        matplotlib.lines.Line2D: The line object created by `ax.plot`.
    """
    normal_vector = np.array(normal_vector)  # Ensure it's a NumPy array for calculations

    if normal_vector.shape != (2,):
        raise ValueError("Normal vector must be a 2D vector (length 2).")

    w1, w2 = normal_vector

    if ax is None:
        fig, ax = plt.subplots()

    if x_range is None:
        # Determine a reasonable x-range
        xmin = -5
        xmax = 5
        x_range = (xmin, xmax)

    xmin, xmax = x_range
    x = np.linspace(xmin, xmax, 100)  # Generate x-values

    if w2 != 0:
        y = (-w1 * x - offset) / w2  # Solve for y: w1*x + w2*y + b = 0  =>  y = (-w1*x - b) / w2
    else:
        # Handle the case where the line is vertical (w2=0). Plot a vertical line.
        if w1 == 0:
            raise ValueError("Cannot plot line with w1=0 and w2=0.")
        x = np.full_like(x, -offset / w1)  # x is constant for a vertical line
        y = np.linspace(ax.get_ylim()[0], ax.get_ylim()[1], 100)  # generate y values spanning the current y-axis.



    line, = ax.plot(x, y, **kwargs)

    return line
```

```python
#part 1: find slope and intercept by using the visualization for the first Class
w1=np.array([2,-1.1])
b1=0

#line1 = plot_line_from_normal_vector(w, b, color='blue',x_range=[0,10])
line1 = plot_line_from_normal_vector_and_offset(w1, b1, color='blue',x_range=[0,6])
plt.scatter(X[:,0],X[:,1],c=y)
plt.quiver(0,b1,w1[0],w1[1])
plt.grid()
```

```python
#part 2: find slope and intercept by using the visualization for the second Class
w2=np.array([2.2,6])
b2=4

#line1 = plot_line_from_normal_vector(w, b, color='blue',x_range=[0,10])
line2 = plot_line_from_normal_vector_and_offset(w2, b2, color='red',x_range=[0,6])
plt.scatter(X[:,0],X[:,1],c=y)
plt.quiver(0,b2,w2[0],w2[1])
plt.grid()
```

```python
#part 3: find slope and intercept by using the visualization for the third Class
w3=np.array([-10,1])
b3=-15


#line1 = plot_line_from_normal_vector(w, b, color='blue',x_range=[0,10])
line3 = plot_line_from_normal_vector_and_offset(w3, b3, color='orange',x_range=[0,6])
plt.scatter(X[:,0],X[:,1],c=y)
plt.quiver(0,b3,w3[0],w3[1])
plt.grid()
```

### Transform lables to one-hot coding
* Hint: Use Scikit-Learn

1. y onehot encoden
2. y skalieren auf -1 und 1
3. y{:,label index} mit np.sign prüfen


```python
# y one Hot encoden
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)  # `sparse_output=False` gibt ein normales NumPy-Array zurück
y_encoded = encoder.fit_transform(y.reshape(-1, 1))
y_encoded
```

```python
# y skalieren
y_scaled=(y_encoded*2-1) #scale true lables from [0,1] to [-1,1
print(y_scaled)
```

### Manually find parameters $w$ and $b$ for
* for each Perceptron seperately oben gemacht

* then combine to single matrix multiplication formulation

```python
# Matrix erstellen
W = np.array([[w1],
              [w2],
              [w3]])

# Bias-Vektor
B = np.array([b1,b2,b3])
```

#### test single perceptrons by class


```python
# erstes Perceptron testen
i=1
print (y_scaled[:,i],np.sign(w1.dot(X[i,:].T)+b1)) # y[:,i]
y[i]==np.sign(w1.dot(X[i,:].T)+b1)
```

```python
# zweites Perceptron testen
i=1
print (y_scaled[:,i],np.sign(w2.dot(X[i,:].T)+b2)) # y[:,i]
y[i]==np.sign(w2.dot(X[i,:].T)+b2)
```

```python
# drittes Perceptron testen
i=1
print (y_scaled[:,i],np.sign(w3.dot(X[i,:].T)+b3)) # y[:,i]
y[i]==np.sign(w3.dot(X[i,:].T)+b3)
```

#### test full Setup

```python
# Alle Perceptrons testen
i=1
print (y_scaled[:,i],np.sign(W.dot(X[i,:].T)+B)) # y[:,i]
y[i]==np.sign(W.dot(X[i,:].T)+B)
```

## Add non-linear activation / loss -> $softmax$
* implement a $softmax$ function
* replace $sgn$ by $softmax$
* plot confidence of the test samples

```python
def softmax(x):
    """Compute softmax values for each sets of scores in x."""
    exp_x = np.exp(x - np.max(x))  # Numerisch stabil
    return exp_x / np.sum(exp_x)

probs = softmax(W) # Matrix Vektor in Funktion implementieren
print(probs)
```

```python
y_final = y_scaled.squeeze()
predictions = np.argmax(W.reshape(3, 2) @ X.T + B[:, np.newaxis], axis=0)  # Reshape W to (3, 2) and adjust prediction calculation

# Convert predictions to one-hot encoding
predictions_onehot = np.zeros_like(y_final)
predictions_onehot[np.arange(len(predictions)), predictions] = 1

# Now compare predictions_onehot with y_final
pos = np.where(np.any(y_final != predictions_onehot, axis=1))
print(pos)
```

```python
y_test=np.copy(y)
y_test[pos]=0
plt.scatter(X[:,0],X[:,1],c=y_test)#set color
plt.colorbar()
#abline(1,2)
```
